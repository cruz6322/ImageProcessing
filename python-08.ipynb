### 클래스 작성

* 입사 지원자의 자소서 관리
* 자소서를 하나의 클래스로 만들고, 각각의 자소서에 대한 객체를 생성

#### 1) 클래스 기본 형태 및 동작

```python
class jss:
    def __init__(self):                       # 클래스의 기본 메소드(객체)로 __init__() 함수 사용
                                              # __init__() 도 함수이므로 입력변수로 self를 필수항목으로 지정 (self는 클래스 자신을 의미)
        print("jss 클래스 선언!!")

    def show(self):                           # 클래스 내에 show() 메소드 생성
        print("show 메소드 실행!")
```

```python
jss()
```

```python
a = jss()             # a 라는 속성(변수) 생성 : a에 jss라는 클래스 자체를 실행 --> self 함수가 실행됨 
```

```python
a.show()                # show()라는 메소드 불러옴
                        # a라는 변수가 jss 클래스를 의미하기때문에, a.show()에는 self 지정이 불필요
```

#### 2) 클래스 작성

```python
class jss:
    def __init__(self):                                  # 클래스의 init()에서는 입력변수로 항상 self 넣음
        self.name = input("이름을 입력하세요 :  ")
        self.age = input("나이를 입력하세요 :  ")

    def show(self):                                      # show() 메소드(객체) 생성
        print("내 이름은 {}, 나이는 {}세 입니다".format(self.name, self.age))       #문자열에 format 함수 사용
```

```python
 a = jss()
```

```python
a.name
#print(a.name)
#print(a.age)
```

```python
a.age
```

```python
a.show()
```

#### 3) 새로운 클래스를 만들어야 할때, 기존 작성된 클래스를 그대로 활용하고자 할 경우
* 기존 클래스를 상속받는 방법을 사용

* 새로운 jss2라는 클래스를 만들고자 하는데, 기존에 만들어진 jss 클래스를 상속받아서 그대로 사용 가능
* 기존 jss 클래스에 성별이라는 속성을 추가하고자 할때

```python
class jss2(jss):                                        # 상속받고자하는 jss를 입력변수로 지정
    def __init__(self):                                 
        super().__init__()                              # super() 함수는 상속받은 jss 클래스를 그대로 사용하겠다는 의미
                                                        # 기존 jss 클래스내의 객체를 그대로 사용할 수 있음
        self.gender = input("성별 입력 :  ")

    def show(self):                                      # show() 메소드(객체) 생성
        print("내 이름은 {}, 성별은 {}성, 나이는 {}세 입니다".format(self.name, self.gender, self.age))       #문자열에 format 함수 사용
```

```python
b = jss2()
```

```python
b.gender
```

```python
b.name
```

```python
b.show()
```
