```python
title = '서기 1년 1월 1일부터 ' \
        '오늘까지 ' \
        '일수 구하기 '                 # 3행에 걸쳐 작성 → 1개 논리적 명령행
months = [31, 28, 31, 30, 31, 30,
          31, 31, 30, 31, 30, 31]
```

```python
title
```

```python
year, month = 2025, 10                  # 여러개 변수를 한개의 논리적 행에 선언
day = 7; ratio = 365.2425               # 2개 논리적 명령행을 세미콜론으로 분리하여
                                        # 한개의 물리적 행으로 표시
```

```python
month, year
```

```python
days = (year -1) * ratio + \
       sum(months[:month-1]) + day        # 수식이 길 경우 줄바꿈 사용
```

```python
days
```

```python
print(title), print(' - 년:', year ), print(' - 월:', month)    # 콤마로 함수 분리
print(' - 일:', day); print(' * 일수 총합:', int(days))         # 세미콜론으로 행 바꿈
```
