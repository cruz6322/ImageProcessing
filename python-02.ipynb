```python
list1 = [1, 2, 3, 4]
list2 = [1, 1.5, 'a', 'a', '문자열']

#print(list1[0])
print('list1 = ', list1, type(list1))
print('list2 = ', list2, type(list2))

list1[0],list1[1] = 5, 6                # list의 원소는 0번부터 시작함
                                        # 0번, 1번 값을 새 값으로 대체
list2.insert(3, 'b')                    # 원소 삽입

print('list2 = ',list2)
```

```python
tuple1 = (1,2)
tuple2 = (1, 1.5, 'b', 'b', '문자열')

# tuple1[0] = 5                         # 에러 발생 - 튜플은 원소 변경불가

print('tuple1 = ', tuple1, type(tuple1))
print('tuple2 = ', tuple2)
tuple2
```

```python
dict1 = { 'name': 'lee', 'email': 'dslee@yuhan.ac.kr' }

print('dict1 = ', dict1, type(dict1))
print('dict1(email) = ', dict1['email'], type(dict1))

dict1['email'] = 'naver.com'            # key에 접근하여 값을 변경

print('dict1 = ', dict1, type(dict1))
```

```python
set1 = set(list2)
set2 = set(tuple2)
set3 = {10, 20.5, 'string'}

print('set1', set1, type(set1))
print('set2', set2, type(set2))
print('set3', set3, type(set3))
#print('\n')
print('insection', set1 & set2 )        # 교집합
```
