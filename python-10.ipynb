### 내장함수 사용

```python
a = [1.5, 2, 3, 4, 5]
b = len(a)
c = divmod(5, 3)
```

```python
print('최대값:', max(a) ,'최소값:' , min(a) )
print("객체의 원소의 갯수 : ", b)
print( '몫과 나머지:' , c )
print( 'c의 자료형:', type(c), type(c[0]), type(c[1]) )
```

```python
print('2의 4제곱:', pow(2, 4) )
print('절댓값:', abs(-4) )
```
