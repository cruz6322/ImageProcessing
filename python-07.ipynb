### 함수 정의 및 사용

```python
# 도형의 면적을 구하는 함수 정의

def calc_area(type, a, b, c = None):    # 4개의 인수를 설정함, 단 c는 초기값으로 none이 설절됨 
    if type == 1 :                      # 사격형
        result = a * b
        msg = '사각형'
    elif type == 2:                     # 삼각형
        result = a * b / 2
        msg = '삼각형'
    elif type == 3:                     # 평행사변형
        result = (a + b) * c / 2
        msg = '평행사변형'
    return result, msg                  # 반환값 – 2개 반환하면 튜플 반환

def say():                              # 단순 프린트 함수 정의 (인수 없음) 
    print ('넓이를 구해요')

def write(result, msg):                # 리턴받은 값을 프린트하는 함수 정의 
    print ( msg,' 넓이는 ', result , '㎡ 입니다.')
```

#### 함수 호출 방법 : 4가지 케이스

```python
# 함수 호출 방법 : 4가지 케이스

say()                                   # 함수 호출 case 1 : 인수&반환값 없음
ret = calc_area(type=1, a=5, b=5)		# 함수 호출 case 2 : 함수 리턴값을 하나의 튜플 형태로 받는 경우, 인수명과 값을 같이 입력
area, msg = calc_area(2, 5, 5)			# 함수 호출 case 3 : 인수명없이 인수값만 입력한 형태, 가장 흔함. c는 초기값을 그대로 입력함. 함수의 리턴값 두개를 반환
area2, _ = calc_area(3, 10, 7, 5)		# 함수 호출 case 4 : 리턴값을 모두 받지 않고, 원하는 반환 값만 지정, 나머지는 언더바로 표시
```

```python

print(type(ret))                   # ret 변수의 클래스는 튜플
print(type(area),type(msg) )       # area의 클래스는 실수형(나눗셈 때문), msg의 클래스는 문자열
write(ret[0], ret[1])              # write 함수의 입력값으로 ret[0]=result, ret[1]=msg 값 입력. ret는 사각형이므로 write 함수 결과값 출력.
write(area, msg)                   # write 함수의 입력값으로 calc_area(2,5,5)를 수행한 반환값 두개를 지정
write(area2, '평행사변형' )        # write 함수의 입력값으로 calc_area(3,10,7.5)를 수행한 반환값 두개를 지정, 두번쩨 반환값으로는 문자를 직접 입력한 형태
```
