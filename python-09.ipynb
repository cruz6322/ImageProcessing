#### 모듈 파일 내의 특정한 함수만 불러오고자 할떼
* from 모듈파일 import 특정함수

```python
import header_area as mod
from header_area import write
```

```python
mod.say()                                       # mode.say() 함수 호출 – 인수없이 호출한 경우, 반환값 없음
ret = mod.calc_area(type=1, a=5, b=5)			# mode.calc_area() 함수 호출 – 튜플 형태로 반환한후, ret 변수에 저장
write(ret[0], ret[1])                           # write 함수에 두 파라미터 입력

#print(ret[0], ret[1])                          # ret 변수에 튜플형태로 2개값이 저장
```
